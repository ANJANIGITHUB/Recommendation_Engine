```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
retail_data=pd.read_excel('OnlineRetail.xlsx')
```

```python
retail_data.head()
```

### Exploratory Data Analysis

```python
retail_data.shape
```

```python
retail_data['Description']=retail_data['Description'].str.strip() #Stripe spaces from Descrition
```

```python
retail_data.dropna(axis=0,subset=['InvoiceNo'],inplace=True)
```

```python
retail_data['InvoiceNo']=retail_data['InvoiceNo'].astype('str')#Convert as string
```

```python
retail_data['InvoiceNo']=retail_data[~retail_data['InvoiceNo'].str.contains('C')] # Remove Credit Transactions
```

```python
retail_data.head()
```

```python
retail_data['Country'].value_counts()
```

```python
#retail_data.groupby(by='Country')['Country'].count()
```

```python
#Separating for Germany
```

```python
mybasket=(retail_data[retail_data['Country']=='Germany'].groupby(['InvoiceNo','Description'])['Quantity'].sum().unstack().reset_index().fillna(0).set_index('InvoiceNo'))
```

```python
mybasket
```

```python
def convert_pos(data):
    if data<=0:
        return 0
    elif data>0:
        return 1
```

```python
mybasket=mybasket.applymap(convert_pos)
```

```python
mybasket
```

```python
mybasket.dtypes
```

### Training Model

```python
my_frequent_items=apriori(mybasket,min_support=0.07,use_colnames=True)
```

```python
my_frequent_items
```

### Recommendation Rules

```python
my_frequent_items_rules=association_rules(my_frequent_items,metric='lift',min_threshold=1)
```

```python
my_frequent_items_rules
```
